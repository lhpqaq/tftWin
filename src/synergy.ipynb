````python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import json
import copy
url = 'https://lol.qq.com/tft/#/synergy'
driver = None
````

````python
class Synergy(object):
    def __init__(self,name,text,buff):
        self.name=name
        self.text=text
        self.champions=[]
        self.buff=buff
class Champion(object):
    def __init__(self,name,cost,synergy):
        self.name=name
        self.cost=cost
        self.sy=synergy
````

一开始这样写的，导致所有实例的champions都是同一个id

````python
def get():
    global driver
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(2)
    sy1=driver.find_elements(By.CLASS_NAME,'synergy-item')
    sy=[]
    time.sleep(2)
    for s in sy1:
        name=s.find_element(By.CLASS_NAME,'synergy').text
        text=s.find_element(By.CLASS_NAME,'txt-p').text
        buf=s.find_elements(By.CLASS_NAME,'txt-p2')
        buff={}
        for i in buf:
            num = int(i.text[0])
            buff[num]=i.text[2:]
        tmp=Synergy(name,text,buff=buff)
        sy.append(tmp)
    driver.find_element(By.XPATH,'//*[@id="appMain"]/div[1]/div/div[1]/a[2]').click()
    sy2=driver.find_elements(By.CLASS_NAME,'synergy-item')
    for s in sy2:
        name=s.find_element(By.CLASS_NAME,'synergy').text
        text=s.find_element(By.CLASS_NAME,'txt-p').text
        buf=s.find_elements(By.CLASS_NAME,'txt-p2')
        buff={}
        for i in buf:
            num = int(i.text[0])
            buff[num]=i.text[2:]
        tmp=Synergy(name,text,buff=buff)
        sy.append(tmp)
    return sy
````

````python
a = get()
b = copy.deepcopy(a)
````

````python
def champ():
    global a,b
    a=copy.deepcopy(b)
    with open("../json/champions.json",'r',encoding='utf-8') as load_f:
        load_dict = json.load(load_f)
    cs = []
    for i in range(len(load_dict)):
        t = load_dict[str(i)]
        name = t['name']
        cost = t['cost']
        sy = t['sy']
        t = Champion(name,cost,sy)
        #print(t.name)
        cs.append(t)
    for c in cs:   #每个英雄
        for s in c.sy:  #单个英雄的每个羁绊
            t=0
            for i in range(len(a)):
                if(a[i].name==s):
                    t=i
                    break
                    #a[i].champions.append(c.name)
            a[t].champions.append(c.name)
````

````python
champ()
````

````python
for i in a:
    print(i.name,end=' : ')
    print(i.champions)
````

````python
dic = {}
for i in range(len(a)):
    dic[i]=a[i].__dict__
````

````python
with open("../json/synergys.json","w",encoding='utf-8') as f:
    json.dump(dic,f)
    print("写入文件完成...")
````

````python
with open("../json/synergys.json",'r',encoding='utf-8') as load_f:
    load_dict = json.load(load_f)
    #print(load_dict)
a = []
for i in range(len(load_dict)):
    t = load_dict[str(i)]
    name = t['name']
    text = t['text']
    buff = t['buff']
    ch = t['champions']
    t = Synergy(name,text,buff)
    t.champions+=ch
    a.append(t)
````
